```python
# What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
#Matplotlib.
```

```python
"""
Ans. Matplotlib is a popular data visualization library in Python that provides a wide range of tools for creating 
     static, animated, and interactive plots
     
     Five plots are: 
     
     1) Line plot
     2) Scatter Plot
     3) Histogram
     4) Bar Plot
     5) Pie Chart
"""
import matplotlib.pyplot as plt

# Line Plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()

# Scatter Plot

a = [1, 2, 3, 4, 5]
b = [2, 4, 6, 8, 10]

plt.scatter(a, b)
plt.xlabel('a-axis')
plt.ylabel('b-axis')
plt.title('Scatter Plot')
plt.show()

# Histogram

data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]

plt.hist(data, bins=5, edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()

# Bar Plot

categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 5]

plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()

# Pie Chart

labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal') 
plt.title('Pie Chart')
plt.show()
```

```python
"""
Ans. Scatter Plot: Scatter plots are used to visualize the relationship between two variables by displaying individual
     data points.
"""

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.scatter(x,y)
plt.title("Question 2")
plt.show()
```

```python
# Why is the subplot() function used? Draw four line plots using the subplot() function.
```

```python
"""
Ans. The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange
     plots in a grid, specifying the number of rows and columns. 
"""
import numpy as np
import matplotlib.pyplot as plt


x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1', color='red')
plt.title('Line Plot 1')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2', color='green')
plt.title('Line Plot 2')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3', color='blue')
plt.title('Line Plot 3')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4', color='purple')
plt.title('Line Plot 4')
plt.legend()


plt.tight_layout()


plt.show()
```

```python
"""
Ans. Bar Plot: Bar plots are effective for comparing values of different categories.
"""
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
"""
Ans. Box Plot: useful for displaying the summary statistics of a dataset, including the median, quartiles, and potential
     outliers.
"""
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Two Datasets')
plt.xlabel('Box')
plt.ylabel('Values')

plt.show()
```
